```python
#### Q4 Generate 1000 2D samples from the standard Gaussian distribution, i.e X ~ N(0,I). Show them in a scatter plot using pyplots.scatter()

NUM_SAMPLES = 1000
mean = np.zeros(2)
cov = np.eye(2)
rand_standard_Gauss_2D = np.random.multivariate_normal(mean,cov, NUM_SAMPLES) #This yields a (sample_size X 2) matrix 
plt.scatter((rand_standard_Gauss_2D[:,0]) , (rand_standard_Gauss_2D[:,1]))
plt.title('1000 2D samples randomly drawn from the standard Gaussian distribution')
plt.xlabel('x components of 1000 2D samples')
plt.ylabel('y componenets of 1000 2D samples')
plt.show()


#### Q5. Multiply those 2D samples by the matrix A = [1 2; 1 1] and plot *them* in a scatter plot.
S1 = 1
S2 = 1
scaleMatrix = np.array([[S1,2],[1,S2]]) ## matrix A 
normalSamples = scaleMatrix@np.transpose(rand_standard_Gauss_2D)


#### Q6. Finally, find the eigenvectors+eigenvalues of the new covariance A*A^T. Using pyplots.arrow(), add to the plot arrows that point in the direction of the eigenvectors and have magnitude equal to the eigenvalues for each eigenvector.
K  = np.dot(scaleMatrix,np.transpose(scaleMatrix)) # new covariance
eigenvalues, eigenvectors = np.linalg.eig(K)
plt.scatter(normalSamples[0,:],normalSamples[1,:])
plt.arrow(0, 0, np.sqrt(eigenvalues[0])*eigenvectors[0,0], np.sqrt(eigenvalues[0])*eigenvectors[1,0], head_width=0.2, head_length=0.1)
plt.arrow(0, 0, np.sqrt(eigenvalues[1])*eigenvectors[0,1], np.sqrt(eigenvalues[1])*eigenvectors[1,1], head_width=0.2, head_length=0.1)
plt.axis([-4,4,-4,4])  #[xmin, xmas,ymin,ymax]
plt.title('1000 2D samples randomly drawn from the standard Gaussian distribution multiplied by the new covariance')
plt.xlabel('x components')
plt.ylabel('y componenets')
plt.show()
```
