# Using Random Numbers and Linear Algebra in NumPy

```python
import numpy as np
import matplotlib.pyplot as plt
```

Generate and display an array of 100 random uniform numbers over [0,1] using numpy.random.uniform().

```python
N = 100
uniform_samples = np.random.uniform(0,1,N)
print(uniform_samples)
```

Use pyplot.subplots() and pyplot.hist() to plot the histogram of the samples using 5, 10, and 20 bins in a 1x3 subplot.

```python
fig, axs = plt.subplots(1,3,sharey=True)
axs[0].hist(uniform_samples,5)
axs[0].set_ylabel("Frequency")

axs[1].hist(uniform_samples,10)
axs[1].set_title("Histogram of Uniform Samples")
axs[1].set_xlabel("Value")

axs[2].hist(uniform_samples,20)

plt.show()
```

Repeat the previous two exercises, but with Gaussian random variables with mean 10 and variance 4

```python
N = 100
gaussian_samples = np.random.normal(10,4,N)
print(gaussian_samples)
```

```python
fig, axs = plt.subplots(1,3,sharey=True)
axs[0].hist(gaussian_samples,5)
axs[0].set_ylabel("Frequency")

axs[1].hist(gaussian_samples,10)
axs[1].set_title("Histogram of Uniform Samples")
axs[1].set_xlabel("Value")

axs[2].hist(gaussian_samples,20)

plt.show()
```

Generate 1000 2D samples from the standard Gaussian distribution, i.e X ~ N(0,I). Show them in a scatter plot using pyplots.scatter()

```python
N = 1000
multivariate_samples = np.random.normal(0,1, [2, N])
plt.scatter(multivariate_samples[0,:],multivariate_samples[1,:])
plt.axis('equal')
plt.show()
```

Multiply those samples by the matrix A = [1 0.5; 0.5 1] and plot *them* in a scatter plot.

```python
A = np.array([[1,2],[1,1]])
transformed_samples = A@multivariate_samples
plt.scatter(transformed_samples[0,:],transformed_samples[1,:])
plt.axis('equal')
plt.show()
```

Finally, find the eigenvectors of the new covariance A&ast;A^T, and add to the plot arrows that point in the direction of the eigenvectors and have magnitude equal to the eigenvalues for each eigenvector.

```python
eigenvalues, eigenvectors = np.linalg.eig(A@A.T)
plt.scatter(transformed_samples[0,:],transformed_samples[1,:])

ax = plt.axes()
ax.arrow(0,0,eigenvalues[0]*eigenvectors[0,0],eigenvalues[0]*eigenvectors[1,0],head_width=0.5, head_length=0.5,fc="k", ec="k")
ax.arrow(0,0,eigenvalues[1]*eigenvectors[0,1],eigenvalues[1]*eigenvectors[1,1],head_width=0.5, head_length=0.5,fc="k", ec="k")

plt.axis('equal')
plt.show()
```
